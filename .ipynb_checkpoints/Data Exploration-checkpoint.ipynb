# ***Data Exploriation***

Data exploration ans visualization are two crtical components of data analysis . They help to understand the underlying strucutre and patterns of the data , which in turn can lead to meaningful insights and decisions.

```python
import pandas as pd 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
# Understanding Descriptive Statistics 
#Descriptive Statistics is the branch of statictics that involves the summarization and describing of data
# pandas provides many build in functions 
```

```python
df.describe()
```

```python
#integrating with other libraries 
import numpy as np 
np_array = df.values
```

```python
np_array
```

```python
import matplotlib.pyplot as plt
df.plot()
plt.show()
```

```python
import seaborn as sb
sb.scatterplot(data=df , x = 'petal_length' , y='petal_width')
plt.show()
```

```python
df.hist(column='petal_length')
plt.show()
```

```python
df.boxplot(column=['petal_length' , 'petal_width' , 'sepal_width' , 'sepal_length'])
```
